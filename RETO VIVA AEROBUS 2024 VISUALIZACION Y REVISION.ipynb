```python
def declare_query_string(**kwargs):

    valid_columns = ['ProductType', 'ProductName', 'Quantity', 'TotalSales', 'UnitPrice']
    query_parts = []

    for col, value in kwargs.items():
        if col in valid_columns and value is not None:
            query_parts.append(f"{col} == '{value}'")

    query_string = " & ".join(query_parts)
    return query_string

# Example usage:
query = declare_query_string(ProductName='Coca Cola', Quantity=100, UnitPrice=10.5)
print(f"Generated query string: {query}")
```

```python
central_stations = ['AO', 'AU', 'AW', 'BA', 'BM', 'AT', 'AD', 'AK']

def add_central_station_columns(dataframe, central_stations):
    
    dataframe['ArrivalCentral'] = dataframe.apply(lambda row: 2 if row['ArrivalStation'] == 'AD' 
                                                  else (1 if row['ArrivalStation'] in central_stations else 0), axis=1)

    dataframe['DepartureCentral'] = dataframe.apply(lambda row: 2 if row['DepartureStation'] == 'AD' 
                                                    else (1 if row['DepartureStation'] in central_stations else 0), axis=1)
    return dataframe


flights = add_central_station_columns(flights, central_stations)
flights.head()
```
